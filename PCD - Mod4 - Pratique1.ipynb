````python
nomes = data.columns.tolist()
tipos = data.dtypes.tolist()

data_tipos = pd.DataFrame( {'Variável': nomes,
                           'Tipo de Dados': tipos} )

data_tipos
````

````python
qtd_categorias = [len(data[i].unique()) for i in nomes]
qtd_categorias
````

````python
data_tipos['Quantidade'] = qtd_categorias

data_tipos
````

````python
qualitativas = data_tipos[data_tipos['Tipo de Dados'].apply( lambda x : x not in [np.number, bool] )]['Variável']
````

````python
data_dummy = pd.get_dummies(data[qualitativas])

data_dummy
````
